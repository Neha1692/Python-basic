Ans- "For" loop is used for iterating over a sequence of elements, while "while" loop is used for executing code until a specified condition becomes false.

```python
#Write a program to demonstrate the use of logical operators.
#logical operator are AND,OR,NOT
a, b, c = True, False, True

# AND: Both conditions must be True
if a and c:
    print("Both a and c are True (AND condition).")

# OR: At least one condition must be True
if b or c:
    print("Either b or c is True (OR condition).")

# NOT: Reverses the condition
if not b:
    print("b is False (NOT condition).")
```

```python
#Write a Python program to find the sum of all even numbers between 1 and 50.
num = int(input('Enter a number: '))
sum = 0
for i in range(0, num+1):
    if i % 2 == 0:
        print(i)
        sum+=i

print(f"Sum of all the even numbers is {sum}")
```
